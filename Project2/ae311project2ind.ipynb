```python
import numpy as np
import matplotlib.pyplot as plt

def camberline(x, p, m):
    z = np.zeros_like(x)
    for i, x_ in enumerate(x):
        if x_ <= p:
            z[i] = (m / p**2) * (2*p*x_ - x_**2)
        else:
            z[i] = (m / (1 - p)**2) * (1 - 2*p + 2*p*x_ - x_**2)
    return z

    
p = 0.3
m_values = np.linspace(0.0, 0.05, 6)


plt.figure(figsize=(10, 3))
for m in m_values:
    x = np.linspace(0.0, 1.0, 100)
    z = camberline(x, p, m)
    plt.plot(x, z, label=f"p={p_}, m={m}")

plt.title("Camber Lines for Different Parameters")
plt.xlabel("Chord-line coordinate (x')(m)")
plt.ylabel("Camber (z')(m)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def camberline(x, p, m):
    z = np.zeros_like(x)
    for i, x_ in enumerate(x):
        if x_ <= p:
            z[i] = (m / p**2) * (2*p*x_ - x_**2)
        else:
            z[i] = (m / (1 - p)**2) * (1 - 2*p + 2*p*x_ - x_**2)
    return z

    
p_ = 0.4
m_values = np.linspace(0.0, 0.05, 6)


plt.figure(figsize=(10, 3))
for m in m_values:
    x = np.linspace(0.0, 1.0, 100)
    z = camberline(x, p, m)
    plt.plot(x, z, label=f"p={p}, m={m}")

plt.title("Camber Lines for Different Parameters")
plt.xlabel("Chord-line coordinate (x') (m)")
plt.ylabel("Camber (z') (m)")
plt.legend()
plt.grid(True)
plt.show()
```
